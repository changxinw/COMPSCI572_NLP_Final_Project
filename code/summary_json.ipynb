```python
datasets = ["IMDB", "TREC", "SST"]
models = ["SingleCNN_static", "SingleCNN_nonstatic", "SingleCNN_spe",
          "kimCNN_static", "kimCNN_nonstatic", "kimCNN_multichannel", 
          "bilstm", "bert"]
```

```python
import pandas as pd
res_param = pd.DataFrame(index = models, columns = datasets)
res_param

inputF = "/hpc/group/jilab/changxin/course/cs572/project/result/"
for m in models:
    if m.startswith("SingleCNN"):
        model_name = "%s_fs3"%m
    else:
        model_name = m
    for d in datasets:
        json_path = "%s/%s_%s.json"%(inputF, model_name, d)
        print(json_path)
        json_dict = json.load(open(json_path))
        res_param.loc[m, d] = json_dict["num_param"]
res_param.to_csv("/hpc/group/jilab/changxin/course/cs572/project/tables/model_number_parameters.csv")
```

```python
import pandas as pd
import statistics
res_time = pd.DataFrame(index = models, columns = datasets)
# res_time
res_time_sd = pd.DataFrame(index = models, columns = datasets)

model_rep = ["model_0", "model_1", "model_2", "model_3", "model_4"]
inputF = "/hpc/group/jilab/changxin/course/cs572/project/result/"
for m in models:
    if m.startswith("SingleCNN"):
        model_name = "%s_fs3"%m
    else:
        model_name = m
    for d in datasets:
        time_vec = []
        json_path = "%s/%s_%s.json"%(inputF, model_name, d)
        json_dict = json.load(open(json_path))
        for modelr in model_rep:
            time_vec = time_vec + json_dict[modelr]["time"]
        res_time.loc[m, d] = statistics.mean(time_vec)
        res_time_sd.loc[m, d] = statistics.stdev(time_vec)
        # res_time.loc[m, d] = json_dict["num_param"]
res_time.to_csv("/hpc/group/jilab/changxin/course/cs572/project/tables/model_mean_time_per_epoch.csv")
res_time_sd.to_csv("/hpc/group/jilab/changxin/course/cs572/project/tables/model_stdev_time_per_epoch.csv")
```

```python
### diferent filter size in SingleCNN group
datasets = ["IMDB", "TREC", "SST"]
models = ["SingleCNN_static", "SingleCNN_nonstatic", "SingleCNN_spe"]
filter_sizes = ["fs2", "fs3", "fs4", "fs6", "fs8", "fs12"]
model_rep = ["model_0", "model_1", "model_2", "model_3", "model_4"]

for d in datasets:
    res_acc = pd.DataFrame(index = models, columns = filter_sizes)
    res_acc_sd = pd.DataFrame(index = models, columns = filter_sizes)
    res_time = pd.DataFrame(index = models, columns = filter_sizes)
    res_time_sd = pd.DataFrame(index = models, columns = filter_sizes)
    res_param = pd.DataFrame(index = models, columns = filter_sizes)
    for m in models:
        for f in filter_sizes:
            json_path = "%s/%s_%s_%s.json"%(inputF, m, f, d)
            json_dict = json.load(open(json_path))
            res_acc.loc[m, f] = statistics.mean(json_dict["test_acc"])
            res_acc_sd.loc[m, f] = statistics.stdev(json_dict["test_acc"])
            res_param.loc[m, f] = json_dict["num_param"]
            time_vec = []
            for modelr in model_rep:
                time_vec = time_vec + json_dict[modelr]["time"]
            res_time.loc[m, f] = statistics.mean(time_vec)
            res_time_sd.loc[m, f] = statistics.stdev(time_vec)
    res_acc.to_csv("/hpc/group/jilab/changxin/course/cs572/project/tables/singleCNN_filtersize_mean_acc_%s.csv" % d)
    res_acc_sd.to_csv("/hpc/group/jilab/changxin/course/cs572/project/tables/singleCNN_filtersize_stdev_acc_%s.csv" % d)
    res_time.to_csv("/hpc/group/jilab/changxin/course/cs572/project/tables/singleCNN_filtersize_mean_time_%s.csv" % d)
    res_time_sd.to_csv("/hpc/group/jilab/changxin/course/cs572/project/tables/singleCNN_filtersize_stdev_time_%s.csv" % d)
    res_param.to_csv("/hpc/group/jilab/changxin/course/cs572/project/tables/singleCNN_filtersize_number_parameters_%s.csv" % d)
```
